```python
from torchmetrics.functional.multimodal import clip_score
from functools import partial

clip_score_fn = partial(clip_score, model_name_or_path="openai/clip-vit-base-patch16")


def calculate_clip_score(images, prompts):
    images_int = (images * 255).astype("uint8")
    clip_score = clip_score_fn(torch.from_numpy(images_int).permute(0, 3, 1, 2), prompts).detach()
    return round(float(clip_score), 4)


sd_clip_score = calculate_clip_score(images, sample_prompts)
print(f"CLIP score: {sd_clip_score}")
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import cv2
import frechet_video_distance as fvd
tf.compat.v1.disable_eager_execution()
# Load real video
real_video_path = '/content/interpolated (2).mp4'
real_video_cap = cv2.VideoCapture(real_video_path)

# Load generated video
generated_video_path = '/content/custom_gen.mp4'
generated_video_cap = cv2.VideoCapture(generated_video_path)

batch_size = 16  # Adjust this value based on your available memory

# Initialize lists to store embeddings
real_embeddings = []
generated_embeddings = []

# Process videos in batches
while True:
    # Extract frames from real video
    real_frames = []
    for _ in range(batch_size):
        ret, frame = real_video_cap.read()
        if not ret:
            break
        real_frames.append(frame)

    # Extract frames from generated video
    generated_frames = []
    for _ in range(batch_size):
        ret, frame = generated_video_cap.read()
        if not ret:
            break
        generated_frames.append(frame)

    # If both videos have no more frames, break out of the loop
    if not real_frames and not generated_frames:
        break

    # Preprocess videos (ensure they are in the range [-1, 1] and have the correct shape)
    real_video = np.array(real_frames)
    real_video = (real_video / 127.5) - 1.0

    generated_video = np.array(generated_frames)
    generated_video = (generated_video / 127.5) - 1.0

    # Create tensors from NumPy arrays
    real_video_tensor = tf.constant(real_video)
    generated_video_tensor = tf.constant(generated_video)

    # Calculate embeddings for the current batch
    with tf.compat.v1.Session() as sess:
        real_embedding = fvd.create_id3_embedding(real_video_tensor)
        generated_embedding = fvd.create_id3_embedding(generated_video_tensor)

    # Append embeddings to the lists
    real_embeddings.append(real_embedding)
    generated_embeddings.append(generated_embedding)

# Concatenate embeddings
real_embeddings = tf.concat(real_embeddings, axis=0)
generated_embeddings = tf.concat(generated_embeddings, axis=0)

# Calculate FVD
fvd_score = fvd.calculate_fvd(real_embeddings, generated_embeddings)

print(f"FVD score: {fvd_score}")
```

```python
import torch
from torch import nn
from torchvision import models, transforms
from PIL import Image
import cv2

# Load pre-trained ResNet-50 model
resnet = models.resnet50(pretrained=True)
resnet.eval()

# Define preprocessing for ResNet-50
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Extract features from a video
def extract_features(video_path):
    features = []
    cap = cv2.VideoCapture(video_path)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess the frame
        frame = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        frame = preprocess(frame)

        # Extract features using ResNet-50
        with torch.no_grad():
            feature = resnet(frame.unsqueeze(0)).flatten()
            features.append(feature)

    cap.release()
    features = torch.stack(features)
    return features

# Calculate Fréchet Video Distance
def calculate_fvd(real_features, generated_features, epsilon=1e-6):
    mu_real, sigma_real = torch.mean(real_features, dim=0), torch.cov(real_features.T)
    mu_generated, sigma_generated = torch.mean(generated_features, dim=0), torch.cov(generated_features.T)

    # Add epsilon to the diagonal of covariance matrices
    sigma_real = sigma_real + epsilon * torch.eye(sigma_real.shape[0])
    sigma_generated = sigma_generated + epsilon * torch.eye(sigma_generated.shape[0])

    # Check for zero covariance matrices
    if torch.any(torch.diag(sigma_real) == 0) or torch.any(torch.diag(sigma_generated) == 0):
        return float('nan')  # Return NaN if zero covariance matrices

    fvd = torch.norm(mu_real - mu_generated) ** 2 + torch.trace(sigma_real + sigma_generated - 2 * torch.sqrt(sigma_real @ sigma_generated))

    return fvd.item()

# Extract features from real and generated videos
real_features = real_features.double()
generated_features = generated_features.double()
fvd_score = calculate_fvd(real_features, generated_features)
print(fvd_score)
```

```python
import tensorflow as tf
import numpy as np
import cv2

# Load the pre-trained Inception V3 model
inception_model = tf.keras.applications.InceptionV3(include_top=False, pooling='avg', input_shape=(299, 299, 3))

# Function to extract features from a video
def extract_features(video_path):
    features = []
    cap = cv2.VideoCapture(video_path)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess the frame
        frame = cv2.resize(frame, (299, 299))
        frame = (frame / 127.5) - 1  # Normalize the frame

        # Extract features using Inception V3
        features.append(inception_model.predict(np.expand_dims(frame, axis=0)))

    cap.release()
    features = np.concatenate(features, axis=0)
    return features

# Function to calculate FVD
def calculate_fvd(real_features, generated_features):
    real_mean, real_cov = tf.nn.moments(real_features, axes=[0])
    generated_mean, generated_cov = tf.nn.moments(generated_features, axes=[0])

    # Ensure covariance matrices have the correct shapes
    real_cov = tf.reshape(real_cov, (1, 1, -1, 1, 1))
    generated_cov = tf.reshape(generated_cov, (1, 1, -1, 1, 1))

    cov_sqrt = tf.linalg.sqrtm(real_cov @ generated_cov)

    fvd = tf.square(tf.linalg.norm(real_mean - generated_mean)) + \
          tf.reduce_sum(real_cov + generated_cov - 2 * cov_sqrt, axis=[-1, -2])

    return fvd

# Extract features from real and generated videos
real_features = extract_features("/content/interpolated (2).mp4")
generated_features = extract_features("/content/custom_gen.mp4")

# Calculate Fréchet Video Distance
fvd_score = calculate_fvd(real_features, generated_features)
print(f"Fréchet Video Distance: {fvd_score}")
```

```python
import cv2
import numpy as np

def histogram_comparison(video1_path, video2_path):
    cap1 = cv2.VideoCapture(video1_path)
    cap2 = cv2.VideoCapture(video2_path)

    total_frames = 0
    total_score = 0

    while (cap1.isOpened() and cap2.isOpened()):
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()

        if not ret1 or not ret2:
            break

        hist1 = cv2.calcHist([frame1], [0], None, [256], [0, 256])
        hist2 = cv2.calcHist([frame2], [0], None, [256], [0, 256])

        comparison_value = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CHISQR)

        total_score += comparison_value
        total_frames += 1

    cap1.release()
    cap2.release()

    if total_frames == 0:
        print("Error: No frames found in one or both videos.")
        return None

    average_score = total_score / total_frames
    return average_score

# Example usage:
video1_path = "/content/custom_gen.mp4"
video2_path = "/content/interpolated (2).mp4"
result = histogram_comparison(video1_path, video2_path)

if result is not None:
    print(f"Average Histogram Comparison Score: {result}")
```

```python
import cv2
import numpy as np

def ssim(video1_path, video2_path):
    cap1 = cv2.VideoCapture(video1_path)
    cap2 = cv2.VideoCapture(video2_path)

    total_frames = 0
    total_score = 0

    while (cap1.isOpened() and cap2.isOpened()):
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()

        if not ret1 or not ret2:
            break

        frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
        frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

        (C1, C2) = cv2.absdiff(frame1, frame2)
        (T1, T2) = cv2.scharr(C1, cv2.CV_32F, 1, 0)
        (T2, T2) = cv2.scharr(C2, cv2.CV_32F, 0, 1)

        ssim_numer = (2 * np.mean(T1 * T2) + 0.01)
        ssim_denom = (np.mean(T1**2) + np.mean(T2**2) + 0.01)
        ssim_value = ssim_numer / ssim_denom

        total_score += ssim_value
        total_frames += 1

    cap1.release()
    cap2.release()

    if total_frames == 0:
        print("Error: No frames found in one or both videos.")
        return None

    average_score = total_score / total_frames
    return average_score

# Example usage:
video1_path = "/content/custom_gen.mp4"
video2_path = "/content/interpolated (2).mp4"
result = ssim(video1_path, video2_path)

if result is not None:
    print(f"Average SSIM Score: {result}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import cv2
import numpy as np

def calculate_fvd(real_features, generated_features):
    real_mean, real_cov = tf.nn.moments(real_features, axes=[0])
    generated_mean, generated_cov = tf.nn.moments(generated_features, axes=[0])

    # Ensure covariance matrices have the correct shapes
    real_cov = tf.reshape(real_cov, (1, 1, -1, 1, 1))
    generated_cov = tf.reshape(generated_cov, (1, 1, -1, 1, 1))

    cov_sqrt = tf.linalg.sqrtm(real_cov @ generated_cov)

    fvd = tf.square(tf.linalg.norm(real_mean - generated_mean)) + \
          tf.reduce_sum(real_cov + generated_cov - 2 * cov_sqrt, axis=[-1, -2])

    return fvd.numpy()  # Convert tensor to NumPy array

def create_fvd_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1, activation='linear'))  # Output layer for FVD score

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def preprocess_video(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, (224, 224))  # Adjust size as needed
        frames.append(frame)

    cap.release()

    return np.array(frames) / 255.0  # Normalize pixel values

def train_fvd_model(model, video1_path, video2_path, epochs=10):
    # Load and preprocess videos
    video1_frames = preprocess_video(video1_path)
    video2_frames = preprocess_video(video2_path)

    # Assuming video frames have shape (num_frames, height, width, channels)
    input_shape = video1_frames.shape[1:]

    # Calculate FVD labels
    fvd_labels = calculate_fvd(video1_frames, video2_frames)

    # Reshape labels to match model expectations
    fvd_labels = np.reshape(fvd_labels, (-1, 1))

    # Train the model
    model.fit(video1_frames, fvd_labels, epochs=epochs, validation_split=0.2)

# Example usage:
video1_path = "/content/custom_gen.mp4"
video2_path = "/content/interpolated (2).mp4"

# Create and train the FVD model
input_shape = (224, 224, 3)  # Adjust based on your video frame size
fvd_model = create_fvd_model(input_shape)
train_fvd_model(fvd_model, video1_path, video2_path)
```

```python
import tensorflow as tf
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.models import Model
import cv2
import numpy as np

# Load the InceptionV3 model pre-trained on ImageNet
def load_inception_model():
    base_model = InceptionV3(weights='imagenet')
    model = Model(inputs=base_model.input, outputs=base_model.get_layer('avg_pool').output)
    return model

# Calculate the Inception Score for a set of frames
def calculate_video_inception_score(frames, inception_model, batch_size=32):
    num_frames = len(frames)

    # Preprocess frames for InceptionV3 model
    frames = preprocess_input(frames)

    # Get InceptionV3 predictions for each frame
    predictions = inception_model.predict(frames, batch_size=batch_size)

    # Calculate Inception Score
    is_score = calculate_inception_score(predictions)

    return is_score

# Calculate the Inception Score for p(y|x)
def calculate_inception_score(p_yx, eps=1E-16):
    # Calculate p(y)
    p_y = np.expand_dims(np.mean(p_yx, axis=0), 0)

    # KL divergence for each frame
    kl_d = p_yx * (np.log(p_yx + eps) - np.log(p_y + eps))

    # Sum over classes
    sum_kl_d = np.sum(kl_d, axis=1)

    # Average over frames
    avg_kl_d = np.mean(sum_kl_d)

    # Handle potential overflow by clipping the value
    clipped_avg_kl_d = np.clip(avg_kl_d, a_min=None, a_max=700)  # 700 is an arbitrary threshold

    # Undo the logs
    is_score = np.exp(clipped_avg_kl_d)

    return is_score

# Example usage:
video_path = "/content/interpolated (6).mp4"

# Read video frames
cap = cv2.VideoCapture(video_path)
frames = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.resize(frame, (299, 299))  # InceptionV3 input size
    frames.append(frame)

cap.release()

# Convert frames to NumPy array
video_frames = np.array(frames)

# Load InceptionV3 model
inception_model = load_inception_model()

# Calculate Inception Score for the video
video_is_score = calculate_video_inception_score(video_frames, inception_model)
print(f"Inception Score for runwayml the video: {video_is_score}")
```
